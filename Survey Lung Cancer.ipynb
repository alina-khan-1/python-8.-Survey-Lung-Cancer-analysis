```python
#how many people have or don't have cancer.
sns.countplot(x="LUNG_CANCER", data = df)

plt.title('Lung Cancer Distribution')
plt.xlabel('Lung Cancer')
plt.ylabel('Count')
plt.show()
df['LUNG_CANCER'].value_counts()
```

##### INSIGHT
270 people in the dataset were diagnosed with lung cancer, while only 39 were not.
This shows a high prevalence of lung cancer in this sample population.

```python
#
sns.histplot(data=df, x='AGE', hue='LUNG_CANCER', bins=10, kde=True)
plt.title('Age Distribution by Lung Cancer Status')
plt.xlabel('Age')
plt.ylabel('Number of People')
plt.show()
```

##### Insight 
Lung cancer appears to be more common in older individuals, especially those aged 6070 years.
This suggests that age is a key risk factor for lung cancer in this dataset.

```python
df.columns
```

```python
sns.countplot(x='SMOKING', hue='LUNG_CANCER', data=df)

plt.title('Smoking and Lung Cancer')
plt.xlabel('Smoking (1=No, 2=Yes)')
plt.ylabel('Count')
plt.show()
df.groupby(['SMOKING', 'LUNG_CANCER']).size()
```

```python
df['SMOKING'].value_counts()
```

#### Insight
More people with Smoking = 2 (Yes) got Lung Cancer(155), compared to those with Smoking = 1 (No)(115).
But also, a large number of non-smokers also got Lung Cancer.

Most people who smoke have lung cancer. But even non-smokers are also getting affected. So, while smoking increases the risk, its not the only reason.

```python
plt.figure(figsize=(12, 8))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

#### Insight

Age is slightly related to coughing (0.17). Maybe older people tend to cough more.

Smoking and anxiety show a small connection (0.16). Smokers may feel more anxious.

Yellow fingers have a strong relation with anxiety (0.57), maybe due to smoking habits.

Anxiety is also linked to yellow fingers  could be signs of unhealthy behavior.

Peer pressure is somewhat related to swallowing difficulty (0.37). People under pressure might face throat issues.

Chronic disease has a weak relation with allergy (0.11). Could mean people with diseases face allergies more often.

Fatigue is related to shortness of breath (0.44). Tired people might have trouble breathing.

Allergy is related to alcohol consumption (0.34). Maybe people with allergies drink more, or vice versa.

Wheezing and coughing show a small connection (0.37). Makes sense since both are respiratory issues.

Alcohol consumption again links back to allergy (0.34), showing repeated connection.

Coughing is related to wheezing (0.37), again showing these symptoms often come together.

Chest pain is related to Alcohol consumption (0.33).

### Conclusion
Based on the analysis:

Lung cancer is highly prevalent in the dataset.

Age and smoking are the most consistent risk factors.

Symptoms like coughing, fatigue, and chest pain are common but not exclusive to lung cancer patients.

Correlations among symptoms (e.g., coughing and wheezing, anxiety and yellow fingers) reveal potential patterns in patient behavior or health conditions.

Alcohol consumption and gender show some influence but are not dominant predictors.


Lung cancer appears to be influenced by a combination of lifestyle factors (like smoking), age, and symptom profiles. However, no single factor guarantees a diagnosis  this reinforces the importance of comprehensive screening and multi-factor risk assessment in healthcare.
